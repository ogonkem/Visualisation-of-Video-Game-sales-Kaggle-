## Visualisation of Video Game console sales

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style("dark")
sns.despine()
```

```python
vg = pd.read_csv('vgsales.csv')
```

```python
vg.head()
```

Each row corresponds to a particular game, the data contains the name of the game, the year it was released as well as a number of categorical traits such as platform, genre and publisher. Also we see that row(game) includes the sales per region; North America, European Union, Japan, other Regions and globally.

### Count unique items in categorical Variable in vg

```python
print("No of games: ", len(vg))
publishers = vg['Publisher'].unique()
print("No of publishers: ", len(publishers))
platforms = vg['Platform'].unique()
print("No of platforms: ", len(platforms))
genres = vg['Genre'].unique()
print("No of genres: ", len(genres))
```

### Check for NULL in vg

```python
vg.isnull().sum()
```

### Dropping all NULL values

```python
vg = vg.dropna()
```

```python
vg.info()
```

```python
vg.isnull().sum()
```

All null values successfully remove

### Create global sales per year bar chart
create a bar chart showing total 'Global_Sales' of video games per year.   
1. Grouping vg by 'Year' and then call .sum() to get total per year. This creates a dataframe with years as the index with the sum of games sales in each region and globally per year.

```python
y = vg.groupby(['Year']).sum()
y
```

notice that the last row above has a Year 2020 which is an impossible release date  
so we remove the that row

```python
vg = vg[vg.Year != 2020.0]
```

```python
y = vg.groupby(['Year']).sum()
y.tail()
```

```python
y = y['Global_Sales']
# also years are floating point numbers eg "2017.0" not "2017". We change the values to integers.
x = y.index.astype(int)
```

```python
plt.figure(figsize=(13,9))
ax = sns.barplot(y = y, x = x)
ax.set_xlabel(xlabel='Year', fontsize=16)
ax.set_xticklabels(labels = x, fontsize=12, rotation=50)
ax.set_ylabel(ylabel='$ Millions', fontsize=16)
ax.set_title(label='Video Game Sales in $ Millions Per Year', fontsize=20)
plt.show();
```

### create bar chart of the total number of releases of video games per year
reate a bar chart showing total 'Global_Sales' of video games per year.  
1. Grouping vg by 'Year' and then call .count() to get total number of releases per year. This creates a dataframe with years as the index or row names and the entries our total sales for that year.


```python
y = vg.groupby(['Year']).count()
y.tail()
```

```python
y = y['Global_Sales']
# also years are floating point numbers eg "2017.0" not "2017". We change the values to integers.
x = y.index.astype(int)
```

```python
plt.figure(figsize=(13,9))
colors = sns.color_palette("muted")
ax = sns.barplot(y = y, x = x)
ax.set_xlabel(xlabel='Year', fontsize=16)
ax.set_xticklabels(labels = x, fontsize=12, rotation=50)
ax.set_ylabel(ylabel='Number of Releases', fontsize=16)
ax.set_title(label='Game Releases Per Year', fontsize=20)
plt.show();
```

```python
platGenre = pd.crosstab(vg.Platform,vg.Genre)
platGenreTotal = platGenre.sum(axis=1).sort_values(ascending = False)
plt.figure(figsize=(13,9))
sns.barplot(y = platGenreTotal.index, x = platGenreTotal.values, orient='h')
plt.ylabel = "Platform"
plt.xlabel = "The amount of games"
plt.show()
```

```python
pub=vg.groupby('Publisher')['Publisher'].count().sort_values(ascending = False).head(20)
plt.figure(figsize=(13,9))
sns.barplot(x=pub.values,y=pub.index)
plt.ylabel = "Top 20 Publishers"
plt.xlabel = "The amount of games"
plt.show()
```

```python
platGenre = pd.crosstab(vg.Genre,vg.Platform)
platGenreTotal = platGenre.sum(axis=1).sort_values(ascending = False)
plt.figure(figsize=(13,9))
sns.barplot(y = platGenreTotal.index, x = platGenreTotal.values, orient='h')
plt.ylabel = "Genre"
plt.xlabel = "The amount of games"
plt.show()
```

```python
table_sales = pd.pivot_table(vg,values=['Global_Sales'],index=['Year'],columns=['Genre'],aggfunc='max',margins=False)

plt.figure(figsize=(19,16))
sns.heatmap(table_sales['Global_Sales'],linewidths=.5,annot=True,vmin=0.01,cmap='PuBu')
plt.title('Max Global_Sales of games', fontsize=16)
```

```python
table_count = pd.pivot_table(vg,values=['Global_Sales'],index=['Year'],columns=['Genre'],aggfunc='count',margins=False)

plt.figure(figsize=(19,16))
sns.heatmap(table_count['Global_Sales'],linewidths=.5,annot=True,fmt='2.0f',vmin=0,cmap='PuBu')
plt.title('Count of games', fontsize=16)
```

```python
GenreGroup = vg.groupby(['Genre']).sum().loc[:, 'NA_Sales':'Global_Sales']
GenreGroup['NA_Sales%'] = GenreGroup['NA_Sales']/GenreGroup['Global_Sales']
GenreGroup['EU_Sales%'] = GenreGroup['EU_Sales']/GenreGroup['Global_Sales']
GenreGroup['JP_Sales%'] = GenreGroup['JP_Sales']/GenreGroup['Global_Sales']
GenreGroup['Other_Sales%'] = GenreGroup['Other_Sales']/GenreGroup['Global_Sales']
plt.figure(figsize=(8, 10))
sns.set(font_scale=0.7)
plt.subplot(211)
sns.heatmap(GenreGroup.loc[:, 'NA_Sales':'Other_Sales'], annot=True, fmt = '.1f',cmap='PuBu')
plt.title("Comparation of each area in each Genre", fontsize=12)
plt.subplot(212)
sns.heatmap(GenreGroup.loc[:,'NA_Sales%':'Other_Sales%'], vmax =1, vmin=0, annot=True, fmt = '.2%',cmap='PuBu')
plt.title("Comparation of each area in each Genre(Pencentage)", fontsize=12)
plt.show()
```
